**Kivy, BeeWare, and frameworks like React Native.**

Kivy is an open-source Python framework for developing multi-touch applications, offering cross-platform compatibility. BeeWare is a collection of tools for developing Python apps that can run on multiple platforms, aiming for native user experiences. React Native is a JavaScript framework used to build mobile applications, leveraging a single codebase for both iOS and Android platforms while maintaining performance through native components.

```python
import http.server
import socketserver

PORT = 8000

Handler = http.server.SimpleHTTPRequestHandler

with socketserver.TCPServer(("", PORT), Handler) as httpd:
    print("Server started at localhost:", PORT)
    httpd.serve_forever()
```

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "Hello, Flask!"

if __name__ == "__main__":
    app.run(debug=True)
```

```python
@app.route("/user/<username>")
def user_profile(username):
    return f"Hello, {username}!"
```

```python
from flask import Flask, render_template

app = Flask(__name__)

@app.route("/")
def home():
    return render_template("index.html", title="Home")

if __name__ == "__main__":
    app.run(debug=True)
```

```python
from flask import Flask, render_template, request

app = Flask(__name__)

@app.route("/contact", methods=["GET", "POST"])
def contact():
    if request.method == "POST":
        name = request.form.get("name")
        message = request.form.get("message")
        return f"Thanks, {name}, for your message: {message}"
    return render_template("contact.html")

if __name__ == "__main__":
    app.run(debug=True)
```
